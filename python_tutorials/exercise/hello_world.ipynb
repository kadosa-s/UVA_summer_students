```python
pts = np.array([(1,2,3,4),(5,6,7,8)])
[x,y]=pts
print([x,y])
```

```python
m,b = np.polyfit(x,y,1)

plt.scatter(x,y)
plt.plot(x, m*x + b)
plt.title('high numbers vs low numbers', fontsize =20)
plt.xlabel('low numbers')
plt.ylabel('high numbers')
plt.show()
```

```python
#second element (starts at 0)
index_remove_x = [1]
#third element (starts at 0)
index_remove_y = [2]

x_reduced = np.delete(x, index_remove_x)
y_reduced = np.delete(y, index_remove_y)
```

```python
# warning: overlapping points
print(x)
print(y)
print(x_reduced)
print(y_reduced)

m,b = np.polyfit(x,y,1)
m_reduced,b_reduced = np.polyfit(x_reduced,y_reduced,1)

plt.scatter(x,y)
plt.scatter(x_reduced,y_reduced)

plt.plot(x, m*x + b)
plt.plot(x_reduced, m*x_reduced + b_reduced)

plt.title('high numbers vs low numbers, orange color deficient arrays', fontsize =20)
plt.xlabel('low numbers')
plt.ylabel('high numbers')
plt.show()
```
